## **Popularity Based Recommender System**
Will Show TOP 40 Books based on its Average Rating , with min 250 rating

```python
# number of rating on each book
book_with_rating=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
book_with_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
book_with_rating
```

```python
# average rating on each book
book_with_avg_rating = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
book_with_avg_rating.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
book_with_avg_rating
```

```python
# merging two average and count table
merge_df = book_with_rating.merge(book_with_avg_rating, on='Book-Title')
merge_df
```

```python
popular_df = merge_df[merge_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(40)
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df
```

```python
# creating amzaon link to pucrchase the book

def generate_link(Book):
    query = Book.replace(" ", "+")
    return f"https://www.amazon.com/s?k={query}"

popular_df['amazon_url'] = popular_df['Book-Title'].apply(generate_link)
```

```python
popular_df
```

##**Collaborate Filtering Based Recommender System**

```python
#keeping only those user who are experience rated atleast 200 books
```

```python
ratings_with_name.head()
```

```python
ratings_with_name.shape
```

```python
y = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
experienced_users=y[y].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(experienced_users)]
```

```python
filtered_rating.shape
```

```python
filtered_rating.head()
```

```python
#filter book which have atleast 50 rating
```

```python
x=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=x[x].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.shape
```

```python
final_rating
```

```python
#creating pivot table
```

```python
final_df=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final_df.shape
```

```python
final_df.fillna(0,inplace=True)
```

```python
final_df
```

```python
# find cosine similarity score of every movie with each other
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_df)
```

```python
similarity_score.shape
```

## **Recommender Function**

```python
#index,and similarity score respect to a book
li=list(enumerate(similarity_score[0]))
```

```python
# sorted the list based on similairyt score and extract the most 4 similar books index
sorted(li,key=lambda x:x[1],reverse=True)[1:6]
```

```python
def recommended(book):
  book_index = np.where(final_df.index==book)[0][0]
  similar_books = sorted(list(enumerate(similarity_score[book_index])),key=lambda x:x[1],reverse=True)[1:7]

  data=[]
  for i in similar_books :
         item=[]
         temp_df=new_book[new_book['Book-Title']==final_df.index[i[0]]]
         temp_df.drop_duplicates('Book-Title')
         item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
         item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
         item.extend(list(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values))
         item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
         item.extend(list(temp_df.drop_duplicates('Book-Title')['amazon_url'].values))
         data.append(item)
  return data

```

```python
recommended('Zoya')
```
